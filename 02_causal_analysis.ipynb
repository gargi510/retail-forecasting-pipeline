## ğŸ› ï¸ Components

### 1. **Exploratory Data Analysis (EDA)**

**Notebook:** `Notebooks/01_EDA_and_Hypothesis_Testing.ipynb`

**Objectives:**
- Data validation of 188,340 training records (Jan 2018 â€“ May 2019) and 22,265 test records (June â€“ July 2019)
- Statistical hypothesis testing using ANOVA and Tukey HSD
- Time series analysis including stationarity and autocorrelation tests
- Distribution and correlation analysis

**Key Findings:**
- **Promotional Impact:** Promotions occur on 45% of days but generate 52% of total sales, increasing AOV by ~7%
- **Holiday Performance:** Sales drop 19% on holidays, indicating defensive (not growth-oriented) discounting
- **Regional Trends:** Region 1 contributes 37% of total sales
- **Data Quality:** Clean dataset with no missing or duplicate values

### 2. **Feature Engineering**

**Module:** `src/features/`

Feature types generated:
- **Lag features:** Historical sales patterns
- **Rolling statistics:** Moving averages and standard deviations
- **Categorical features:** Store, item, region encodings
- **Date features:** Day of week, month, seasonality indicators
- **Promotion features:** Discount depth, promotion duration

All features are generated using **leak-safe pipelines** to prevent data leakage.

### 3. **Causal Analysis**

**Notebook:** `Notebooks/02_causal_analysis.ipynb`

- Measures true promotional uplift using causal inference
- Residual analysis to identify incremental impact
- Counterfactual modeling to estimate sales without promotions

### 4. **Predictive Modeling**

#### **XGBoost Model**
- Gradient boosting with hyperparameter tuning (Optuna)
- Validation SMAPE: **~5.90%**
- Script: `src/training/train_xgb.py`

#### **Neural Network Model**
- Feedforward architecture with batch normalization
- RobustScaler for features, StandardScaler for target
- Validation SMAPE: **~5.31%**
- Script: `src/training/train_nn.py`

#### **Ensemble Model**
- Weighted combination of XGBoost + Neural Network
- Improves stability and reduces bias
- Final Validation SMAPE: **~5.83%**
- Script: `src/ensemble.py`

### 5. **Validation & Metrics**

**Script:** `src/validate.py`

Metrics calculated:
- **SMAPE** (Symmetric Mean Absolute Percentage Error)
- **RMSE** (Root Mean Square Error)
- Model disagreement statistics
- Bias summaries

````python
# ==========================================================
# 04_causal_analysis.py â€” Production-Ready
# ==========================================================
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
import json
import os

# ==========================================================
# 1. DATA PREPARATION
# ==========================================================
train_path = "../data/processed/processed_train_v1.parquet"
train = pd.read_parquet(train_path)

# Ensure Sales_Capped exists
train['Sales_Capped'] = train.get('Sales_Capped', train['Sales'])

# Filter analysis period (Jan-May 2019)
did_df = train[(train['Date'] >= '2019-01-01') & (train['Date'] <= '2019-05-31')].copy()

# ==========================================================
# 2. FEATURE ENGINEERING FOR CAUSAL ANALYSIS
# ==========================================================
did_df['is_promo'] = (did_df['Discount'] == 'Yes').astype(int)
did_df['is_friday'] = (did_df['Date'].dt.dayofweek == 4).astype(int)
did_df['month'] = did_df['Date'].dt.month

# ==========================================================
# 3. FIXED EFFECTS REGRESSION MODEL
# ==========================================================
formula = "Sales_Capped ~ is_promo + is_friday + C(Store_Type) + C(Region_Code) + C(month)"
model = smf.ols(formula=formula, data=did_df).fit()

# ==========================================================
# 4. EXTRACT RESULTS
# ==========================================================
causal_lift_val = model.params.get('is_promo', 0.0)
friday_impact = model.params.get('is_friday', 0.0)
avg_sales_no_promo = did_df.loc[did_df['is_promo'] == 0, 'Sales_Capped'].mean()
perc_lift = (causal_lift_val / avg_sales_no_promo) * 100 if avg_sales_no_promo else 0.0

print(f"--- Causal Analysis Results ---")
print(f"Incremental Sales per Promo Day: ${causal_lift_val:.2f}")
print(f"Causal Promotion Lift: {perc_lift:.2f}%")
print(f"Friday Performance Delta: ${friday_impact:.2f}")

# ==========================================================
# 5. VISUAL VALIDATION & IMAGE EXPORT
# ==========================================================
# Create images folder if it doesn't exist
images_dir = '../models/artifacts/images'
os.makedirs(images_dir, exist_ok=True)

# 5a. March Promo vs Non-Promo Trends
march_trends = (
    did_df[did_df['month'] == 3]
    .groupby(['Date', 'is_promo'])['Sales_Capped']
    .mean()
    .reset_index()
)

march_plot_path = os.path.join(images_dir, 'march_promo_vs_nonpromo.png')
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=march_trends,
    x='Date',
    y='Sales_Capped',
    hue='is_promo',
    palette='viridis',
    marker='o'
)
plt.title("March 2019: Sales Behavior (Promo vs Non-Promo Days)")
plt.ylabel("Avg Sales ($)")
plt.xlabel("Date")
plt.grid(True, alpha=0.3)
plt.legend(title='Promo Day')
plt.tight_layout()


# ==========================================================
# 6 Residuals Plot
# ==========================================================
residual_plot_path = os.path.join(images_dir, 'causal_residuals.png')
plt.figure(figsize=(10, 4))
plt.scatter(did_df['Date'], did_df['causal_residuals'], alpha=0.1, s=1)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residual Analysis: Unexplained Sales Volatility")
plt.xlabel("Date")
plt.ylabel("Residuals")
plt.grid(True, alpha=0.3)
plt.tight_layout()

````

# Key Observations & Business Insights

1. **True Promotion Power**  
   - **Observation:** Causal Lift of 30.83% (slightly below raw 32% uplift).  
   - **Insight:** Most of the sales increase is truly driven by promotions; only ~1.2% was organic.

2. **Friday Anomaly (The Defensive Day)**  
   - **Observation:** Promotions on Fridays reduce sales by \$1,414.54 on average.  
   - **Insight:** Friday discounts are less effective and mostly maintain customer engagement rather than drive new sales.  
   - **Recommendation:** Shift aggressive â€œFlash Salesâ€ mid-week (Tuesdayâ€“Thursday) to maximize incremental lift.

3. **Market Resilience During Volatile Months**  
   - **Observation:** Despite a -20% MoM decline in March, causal lift remained stable.  
   - **Insight:** Promotions are effective even during downturns, generating roughly \$11.6k daily incremental revenue.

4. **Residual/Error Insights**  
   - **Weekly Variance:** Day-of-week volatility remains; promotions may have different â€œdecay ratesâ€ across weekdays.  
   - **Extreme Negative Spikes (Late March/Early April):** Some stores saw sharper-than-average drops, indicating causal lift may be optimistic in severe dips.  
   - **Positive Residual Expansion (May Recovery):** Post-March rebound shows sales exceeding the causal baseline, suggesting momentum effects that the linear model misses.  

**Executive Takeaway:**  
While the Fixed Effects model captures the average lift (\$11,590/day), ~40% of daily volatility remains unexplained. Short-term pulses, extreme market shifts, and rebound effects will require non-linear models (e.g., XGBoost with rolling averages and weekday features) to fully capture.
